```python
batch_size =255
model = Sequential()
model.add(LSTM(30, batch_input_shape=(batch_size,5,4), stateful=True,return_sequences=True,activation='tanh',))
model.add(Dropout(0.4))
model.add(LSTM(units=20,activation='tanh'))
model.add(Dropout(0.4))
model.add(Dense(20,activation='tanh'))
model.add(Dropout(0.3))
model.add(Dense(10,activation='tanh'))
model.add(Dropout(0.3))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')
model.fit(trainX, trainY, epochs=80, batch_size=batch_size, verbose=0, shuffle=False)
model.reset_states()  
pt=model.predict(trainX, batch_size=batch_size)
mp=np.mean(pt[:,0])
print(mp)
def wilo(start,end):
    sum=[]
    for i in range(start,end):
        if(p[i]>mp+0.03):
            sum.append(test[i])
        elif(p[i]<(mp-0.03)):sum.append(-test[i])
    return sum
def wilo1(start,end):
    sum=[]
    for i in range(start,end):
        if(p[i]>0):
            sum.append(test[i])
        elif(p[i]<0):sum.append(-test[i])
    return sum
p=model.predict(testX, batch_size=batch_size)
plt.plot(p)
plt.show()
re=wilo(0,len(testX)-1)
print(np.mean(re),sum(re))
ree=wilo1(0,len(testX)-1)
print(np.mean(ree),sum(ree))
print(sum(test))
print(np.corrcoef(trainY[:,0],pt[:,0]))
print(np.corrcoef(testY[6:,0],p[:-1,0]))
commu=[0]
commuu=[0]
for i in range(0,254):
    x=sum(re[:i])
    y=sum(ree[:i])
    commu.append(x)
    commuu.append(y)
plt.plot(commu)
plt.show()
plt.plot(commuu)
plt.show()
```
