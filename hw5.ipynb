I am interested in how age effect a player's salary

```python
right=pd.read_sql("SELECT playerID,yearID,salary FROM Salaries Where yearID BETWEEN 2006 AND 2016;", db)#table from Salary
left=pd.read_sql("SELECT playerID, birthYear FROM People Where birthYear > 1970 ORDER BY playerID;", db)#get age from Player table
tbe=pd.merge(left, right, on='playerID', how='outer')#combine tables
tb=tbe.dropna()# drop NAN
tb['age']=tb['yearID']-tb['birthYear']#add age
sns.lineplot("age","salary",data=tb)#plot age to salary
```
